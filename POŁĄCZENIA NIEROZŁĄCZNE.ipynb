# POŁĄCZENIA NIEROZŁĄCZNE

Dwa płaskowniki ze stali St2S o grubości g=15mm połączono spoiną czołową. Obciążono stałą siłą rozciągającą. 100kN. Obliczyć szerokość płaskownika (spoiny), niezbędną do przeniesienia danego obciążenia.

krst = kr * s

krst - naprężenia dopuszczalne dla spoiny obciążonej statycznie

kr - naprężenia dopuszczalne dla materiału

s - współczynnik statycznej wytrzymałości spoiny (zależny od rodzaju spoiny i charakteru obciążenia)

Odczytujemy naprężenia dopuszczalne dla stali St2S (wartość naprężeń dopuszczalnych dla materiału)


```python
kr=115000000 #115 MPa = 115000000 Pa
```

Tablica 2. Wartość współczynnika statycznej wytrzymałości spoin s (w zależności od rodzaju spoiny i charakteru obciążenia):

•	dla spoin czołowych przy ściskaniu osiowym i zginaniu s = 1,

•	dla spoin czołowych przy rozciąganiu osiowym i przy zginaniu s = 0.8 ÷ 1, a gdy spoiny są sprawdzane metodą radiograficzną lub ultradźwiękową s = 1,

•	dla spoin czołowych przy ścinaniu s = 0.6,

•	dla spoin pachwinowych niezależnie od charakteru obciążenia s = 0.65.


```python
s=0.8
```

```python
krst = kr*s
"Naprężenia dopuszczalne dla spoiny obciążonej statycznie krst wynosi " + str(krst) + " [N/m2]"
```

```python
F=110000 #100kN = 110000N
```

As = a * l

a = g (g- grubość cieńszego z elementów)

As = g * l 

As = 0.015m * l

```python
a = g = 0.015
```

F/As ≤ krst 

F/(a*l) ≤ krst

Poszukujemy minimalnej wartości l (szerokości płaskownika równej długości spoiny)

F/(a*l) = krst

Po przekształceniu:

l = F/(a*krst)

```python
l = F/(a*krst)
"Przyjmujemy l = " + str(l) +" metra czyli " + str(l*1000) +" milimetra"
```

Uwzględniając, „początek” i „koniec” spoiny

```python
lrzecz = l + 2*a
```

```python
"Minimalna szerokość płaskownika (długość spoiny) wynosi "+ str(lrzecz) +" metra, czyli "+ str(lrzecz*1000) +" milimetra"
```
